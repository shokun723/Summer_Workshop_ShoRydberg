```python
import os
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
#---------------------------
# DATA IMPORTING
#---------------------------



# Relative path to the folder
path = "AI-semicond-workshop/AI-semicond-workshop/device 2"

# List all PNG files
files = sorted([f for f in os.listdir(path) if f.endswith('.png')])
print("Images found:", files)

# Load images and get their shapes
images = []
shapes = []
modes = []
for f in files:
    img = Image.open(os.path.join(path, f))
    images.append(img)
    shapes.append(img.size)
    modes.append(img.mode)
    print(f"{f}: shape {img.size}, mode {img.mode}")

# Preview a user-defined image (change preview_index to select which one, 0-based)
preview_index = 0  # e.g., 0 for first image, 1 for second, etc.
if images:
    plt.imshow(images[preview_index])
    plt.title(f"Preview: {files[preview_index]}")
    plt.axis('off')
    plt.show()
else:
    print("No images found.")

print("pixels", np.array(images[0]).shape)
```

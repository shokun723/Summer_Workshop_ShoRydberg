Use this notebook as your laboratory notebook for the workshop. Work through the lecture notebooks in order, then record observations, results, questions, and evidence here.

A good lab notebook is not just a place for final answers. It should show what you tried, what you noticed, what confused you, and how your thinking changed.

## How to use this notebook

For each lecture notebook:

1. Read the purpose before running the notebook.
2. Make a prediction before major results appear.
3. Run the lecture notebook section by section.
4. Copy important numbers, screenshots, plots, or short notes into this log.
5. Answer the comprehension questions in complete thoughts.
6. Write down one question you would ask a teammate, instructor, or future version of yourself.

## Lab notebook habits

- Date each lab entry.
- Record settings you changed.
- Record results even when they are surprising or messy.
- Keep evidence close to your conclusion.
- Do not erase mistakes. Briefly explain what happened and what you changed.
